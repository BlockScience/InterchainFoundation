# Testing the Bonding Curve Mechanisms

We test the functionality of the bonding curve and its constituent mechanisms, **bond-to-mint** and **burn-to-withdraw**.

## Bond-to-Mint

An agent employs a bond-to-mint mechanism to bond $\Delta R$ quantity of reserve tokens to mint $\Delta S$ quantity of supply tokens.


<div align="center">
	<img width="300" src="img/bond.png" alt="bondtomint">
	<br>
</div>

## Burn-to-Withdraw

An agent employs a burn-to-withdraw mechanism to burn $\Delta S$ quantity of supply tokens to withdraw $\Delta R$ quantity of reserve tokens.


<div align="center">
	<img width="300" src="img/burn.png" alt="burntowithdraw">
	<br>
</div>

These two mechansims work in conjunction to allow agents to trade their reserve and supply tokens in the bonding curve. 

In order to test the bonding curve in isolation, we remove all partial state update blocks (PSUBs) that resolve for the prediction market mechanisms, and only maintain the PSUBs that facilitate trading in the bonding curve. 

```python
import sys
sys.path.append('../')

from src.sim import run
import matplotlib.pyplot as plt
```

```python
experiments = run.run()
```

### Kappa

Expectation: $\kappa$ does not change under the bond-to-mint and burn-to-withdraw mechanism.

Result

```python
plt.plot(experiments.dataset[0].kappa)
plt.title('Kappa')
plt.ylabel('$\kappa$')
plt.xlabel('t')
```

### Alpha

Expectation: $\alpha$ does not change under the bond-to-mint and burn-to-withdraw mechanism.

Result

```python
plt.plot(experiments.dataset[0].alpha)
plt.title('Alpha')
plt.ylabel('alpha')
plt.xlabel('t')
```

### Reserve

Expectation: Reserve varies in accordance to update_R() function as the agents execute bond-to-mint and burn-to-withdraw trades. 

Result

```python
plt.plot(experiments.dataset[0].reserve)
plt.title('Reserve')
plt.ylabel('R')
plt.xlabel('t')
```

### Supply

Expectation: Supply varies in accordance to update_S() function as the agents execute bond-to-mint and burn-to-withdraw trades. 

Result

```python
plt.plot(experiments.dataset[0].supply)
plt.title('Supply')
plt.ylabel('S')
plt.xlabel('t')
```

### Price

Expectation: Price varies in accordance to update_P() function as the agents execute bond-to-mint and burn-to-withdraw trades. 
    
Result

```python
plt.plot(experiments.dataset[0].price)
plt.title('Price')
plt.ylabel('P')
plt.xlabel('t')
```

### Spot Price

Expectation: Spot Price varies in accordance to update_P_bondburn() function as the agents execute bond-to-mint and burn-to-withdraw trades. 
    
Result

```python
plt.plot(experiments.dataset[0].spot_price)
plt.title('Spot Price')
plt.ylabel('Spot Price')
plt.xlabel('t')
```

### Invariant V

Expectation: Invariant V is invariant on the bonding curve, so it should not change.

Result

```python
plt.plot(experiments.dataset[0].invariant_V)
plt.title('Invariant V')
plt.ylabel('V')
plt.xlabel('t')
```

### Invariant I

Expectation: The indifference invariant, I is invariant on the prediction market attestation mechansims but does vary upon the bonding curve bond-to-mint and burn-to-withdraw mechansims.

Result

```python
plt.plot(experiments.dataset[0].invariant_I)
plt.title('Invariant I')
plt.ylabel('I')
plt.xlabel('t')
```

### Price vs Spot Price

```python
plt.plot(experiments.dataset[0].price)
plt.plot(experiments.dataset[0].spot_price)
plt.title('Price and Spot Price')
plt.ylabel('Price, Spot Price')
plt.xlabel('t')
```

### Difference between Spot Price and Price

Expectation: Since the update functions for both spot price and price are similar, we expect no difference between Spot Price and Price. 

Result

```python
plt.plot(experiments.dataset[0].price-experiments.dataset[0].spot_price)
plt.title('Difference between Spot Price and Price')
plt.ylabel('Price - Spot Price')
plt.xlabel('t')
```

The difference between Spot Price and Price is in the order of magnitude $10^{-3}$. This could be caused due to lack of computational precision and rounding errors.

### Difference between Supply and Reserve

Expectation: We expect to see a difference between Supply and Reserve guided by the conservation function with invariant V. $$V = \frac{S^{\kappa}}{R}$$

Result

```python
plt.plot(experiments.dataset[0].supply - experiments.dataset[0].reserve)
plt.title('Difference between Supply and Reserve')
plt.ylabel('S - R')
plt.xlabel('t')
```

### Supply:Reserve Ratio

Expectation: The ratio between Supply and Reserve varies according to the conservation function with invariant V. $$V = \frac{S^{\kappa}}{R}$$ which gives $$VS^{1-\kappa} = 1$$

Result

```python
plt.plot(experiments.dataset[0].supply / experiments.dataset[0].reserve)
plt.title('Supply:Reserve Ratio')
plt.ylabel('Supply:Reserve')
plt.xlabel('t')
```

Plot $VS^{1-\kappa}$

```python
plt.plot((experiments.dataset[0].invariant_V*(experiments.dataset[0].supply)**(1-experiments.dataset[0].kappa)))
plt.title('$VS^{1-\kappa}$')
plt.ylabel('$VS^{1-\kappa}$')
plt.xlabel('t')
```

### Computed Kappa

Expectation: $\kappa$ does not change under the bonding curve bond-to-mint and burn-to-withdraw mechanisms.
    
Result

```python
plt.plot(experiments.dataset[0].spot_price * experiments.dataset[0].supply  / experiments.dataset[0].reserve)
plt.title('Computed Kappa')
plt.ylabel('$\kappa$')
plt.xlabel('t')
```

The computed kappa strays from its initial value of 2 by an order of magnitude $10^{-3}$. This could be caused due to lack of computational precision and rounding errors.

Plotting the difference between the initial kappa value 2 and the computed kappa on a log scale, we have

```python
plt.plot(2-(experiments.dataset[0].spot_price * experiments.dataset[0].supply  / experiments.dataset[0].reserve))
plt.yscale('log')
```

```python
plt.plot(experiments.dataset[0].kappa-(experiments.dataset[0].spot_price * experiments.dataset[0].supply  / experiments.dataset[0].reserve))
plt.yscale('log')
```
