# Private Alpha's dependence on Public Alpha

## Test Case 1: High dependence on public alpha

We initialize the model with 20 agents. All agents start with an initial private beleif of alpha of 0.5 and inital private belief of price of 1.5. All agents have a high bias towards public alpha, with bias coefficient $b = 0.8$. This corresponds to a high dependence of their private belief to the public alpha signal received by interpreting on-chain events.  

$$\hat\alpha_{t = 0} = 0.5$$
$$\hat p_{t = 0} = 1.5$$
$$ b = 0.8 $$

```python
import pandas as pd

import matplotlib.pyplot as plt
# import run2

from src.sim import run

import seaborn as sns

# For analysis
import numpy as np
import pandas as pd

sns.set_style("whitegrid")

experiments = run.run()
```

```python
print(experiments.dataset[0].alpha)
plt.plot(experiments.dataset[0].alpha)
```

```python
print(experiments.dataset[0].kappa)
plt.plot(experiments.dataset[0].kappa)
```
